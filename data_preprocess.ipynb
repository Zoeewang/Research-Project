### document create

```python
# https://stackoverflow.com/questions/38127209/how-to-use-groupby-to-concatenate-strings-in-python-pandas
document = df.groupby('Doc')['content'].apply(','.join)
document
```

```python
document = document.reset_index()
```

```python
document.columns = ['id','text']
document
```

```python
document[['date','speaker']] = document['id'].str.split('-',expand=True)
document
```

```python
document['fed'] = None
document
```

## Add Fed(YES or NO) to df

```python
for index, row in document.iterrows():
    for i, r in biodata.iterrows():
        if row["speaker"]==r["name"]:
            row["fed"]=r['d_regional_fed']
document
```

```python
for index, row in document.iterrows():
    if row["fed"] is None:
        row["fed"]="NA"
document
```

```python
document['change'] = None
document
```

## Add change decision(a,b,c,d,e) to df

```python
for index, row in document.iterrows():
    for i, r in changedata.iterrows():
        if row["date"]==r["start_date"]:
            row["change"]=r['change']
document
```

```python
for index, row in document.iterrows():
    if row["change"] is None:
        row["change"]="NA"
document
```
