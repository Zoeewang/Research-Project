```python
import pandas as pd 
data = pd.read_csv("FOMC_token_separated_col.csv") 
data
```

```python
biodata = pd.read_csv("biographydata.csv")
biodata
```

```python
data['Date'] = data['Date'].astype(str)
```

```python
data["Doc"] = data[['Date', 'Speaker']].agg('-'.join, axis=1)
data
```

```python
df=data[['Doc','content','Speaker']]
df['Doc'] = df['Doc'].astype('str') 
df['content'] = df['content'].astype('str') 
df
```

```python
# https://stackoverflow.com/questions/38127209/how-to-use-groupby-to-concatenate-strings-in-python-pandas
document = df.groupby('Doc')['content'].apply(','.join)
document
```

```python
document = document.reset_index()
```

```python
document.columns = ['id','text']
document
```

```python
document[['date','speaker']] = document['id'].str.split('-',expand=True)
document
```

```python
document['fed'] = None
document
```

```python
for index, row in document.iterrows():
    for i, r in biodata.iterrows():
        if row["speaker"]==r["nameupper"]:
            row["fed"]=r['d_regional_fed']
document
```
