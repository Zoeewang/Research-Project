```python
import pandas as pd 
data = pd.read_csv("FOMC_token_separated_col.csv") 
data
```

```python
data['Date'] = data['Date'].astype(str)
```

```python
data.dtypes
```

```python
data["Doc"] = data["Date"] + data["Speaker"]
data
```

```python
df=data[['Doc','content']]
df['Doc'] = df['Doc'].astype('str') 
df['content'] = df['content'].astype('str') 
df.dtypes
```

```python
# https://stackoverflow.com/questions/38127209/how-to-use-groupby-to-concatenate-strings-in-python-pandas
document = df.groupby('Doc')['content'].apply(','.join)
document
```

```python
document = document.reset_index()
```

```python

document.columns = ['id','text']
document
```

```python
document.to_json('speaker_doc.jsonlist',orient='records', lines=True)
```

```python
document.to_csv("speaker_doc.csv")
```

```python
import csv
import json

with open('speaker_doc.csv') as f:
    reader = csv.DictReader(f)
    rows = list(reader)

with open('speaker_doc.json', 'w') as f:
    json.dump(rows,f)
```
